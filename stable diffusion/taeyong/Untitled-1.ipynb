```python
# Importing Required Modules
from rembg import remove
from PIL import Image
#   (morph_erode.py)

import cv2
import numpy as np

```

```python


# Store path of the image in the variable input_path
input_path = "./datas/sohee1.png"
# Store path of the output image in the variable output_path
output_path = "./datas/detected_sohee1.png"

# Processing the image
input = Image.open(input_path)

# Removing the background from the given Image
output = remove(input)
```

```python
output
```

```python
output.save("./onlynookee.png")
```

```python


width, height = output.size
```

```python

for x in range(width):
    for y in range(height):
        r = output.getpixel((x, y))
        #  (0, 0, 0)   ( )
        if r == (0, 0, 0, 0):
            output.putpixel((x, y), (0, 0, 0))
        else:
            output.putpixel((x, y), (255, 255, 255))
```

```python


output.save(output_path)

img = cv2.imread(output_path)
#   ,  (3x3)  ---
k = cv2.getStructuringElement(cv2.MORPH_RECT, (10,10))
#    ---
dilation = cv2.dilate(img, k)

cv2.imwrite(output_path, dilation)
```
